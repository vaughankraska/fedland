```python
def get_json(file_path):
    with open(file_path) as f:
        data = json.load(f)
    return data
```

```python
def get_experiment_data_5clients(file_path):
    import numpy as np
    import pandas as pd
    training_data0 = get_json(file_path+'/0/training.json')
    training_data1 = get_json(file_path+'/1/training.json')
    training_data2 = get_json(file_path+'/2/training.json')
    training_data3 = get_json(file_path+'/3/training.json')
    training_data4 = get_json(file_path+'/4/training.json')
    validate_data0 = get_json(file_path+'/0/validate.json')
    validate_data1 = get_json(file_path+'/1/validate.json')
    validate_data2 = get_json(file_path+'/2/validate.json')
    validate_data3 = get_json(file_path+'/3/validate.json')
    validate_data4 = get_json(file_path+'/4/validate.json')
    training_data0 = pd.DataFrame(training_data0)
    training_data1 = pd.DataFrame(training_data1)
    training_data2 = pd.DataFrame(training_data2)
    training_data3 = pd.DataFrame(training_data3)
    training_data4 = pd.DataFrame(training_data4)
    validate_data0 = pd.DataFrame(validate_data0)
    validate_data1 = pd.DataFrame(validate_data1)
    validate_data2 = pd.DataFrame(validate_data2)
    validate_data3 = pd.DataFrame(validate_data3)
    validate_data4 = pd.DataFrame(validate_data4)
    return training_data0, training_data1, training_data2, training_data3, training_data4, validate_data0, validate_data1, validate_data2, validate_data3, validate_data4
```
